Task 2 - User Engagement Analysis

Task 2.1 - Part 1: Aggregate Metrics per Customer (MSISDN)

```python
import pandas as pd

def aggregate_user_engagement(df_cleaned):
    
    session_frequency = df.groupby('MSISDN/Number')['MSISDN/Number'].count().rename('session_frequency')

  
    session_duration = df.groupby('MSISDN/Number')['Dur. (ms)'].sum().rename('total_duration')

   
    total_traffic = df.groupby('MSISDN/Number')[['Total UL (Bytes)', 'Total DL (Bytes)']].sum()
    total_traffic['total_traffic'] = total_traffic['Total UL (Bytes)'] + total_traffic['Total DL (Bytes)']

  
    engagement_metrics = pd.concat([session_frequency, session_duration, total_traffic['total_traffic']], axis=1)
    
    return engagement_metrics

# Get the engagement metrics
engagement_metrics = aggregate_user_engagement(df_cleaned)


display(engagement_metrics.head())

# Top 10 users per engagement metric
top_10_session_frequency = engagement_metrics.nlargest(10, 'session_frequency')
top_10_session_duration = engagement_metrics.nlargest(10, 'total_duration')
top_10_total_traffic = engagement_metrics.nlargest(10, 'total_traffic')


print("Top 10 users by session frequency:")
display(top_10_session_frequency)

print("\nTop 10 users by session duration:")
display(top_10_session_duration)

print("\nTop 10 users by total traffic:")
display(top_10_total_traffic)
```

Task 2.1 - Part 2: Normalize Metrics and K-Means Clustering

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


scaler = MinMaxScaler()
normalized_metrics = scaler.fit_transform(engagement_metrics)

# Run K-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['cluster'] = kmeans.fit_predict(normalized_metrics)


display(engagement_metrics.head())

cluster_stats = engagement_metrics.groupby('cluster').agg({
    'session_frequency': ['min', 'max', 'mean', 'sum'],
    'total_duration': ['min', 'max', 'mean', 'sum'],
    'total_traffic': ['min', 'max', 'mean', 'sum']
})


display(cluster_stats)


cluster_stats.plot(kind='bar', figsize=(10, 6))
plt.title('Cluster Statistics for Engagement Metrics')
plt.ylabel('Metric Value')
plt.xlabel('Cluster')
plt.show()
```

Task 2.1 - Part 3: Aggregate User Traffic per Application

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define columns_to_use at the module level
columns_to_use = [
    'Youtube DL (Bytes)',
    'Youtube UL (Bytes)', 
    'Netflix DL (Bytes)',
    'Netflix UL (Bytes)',
    'Google DL (Bytes)',
    'Google UL (Bytes)',
    'Email DL (Bytes)',
    'Email UL (Bytes)',
    'Gaming DL (Bytes)',
    'Gaming UL (Bytes)',
    'Other DL (Bytes)',
    'Other UL (Bytes)'
]

def aggregate_user_traffic_per_application(df_cleaned):
    app_traffic = df_cleaned.groupby('MSISDN/Number')[columns_to_use].sum()
    return app_traffic

user_app_traffic = aggregate_user_traffic_per_application(df_cleaned)

# Find the top 3 applications by maximum traffic across all categories
max_traffic = user_app_traffic.max()
top_3_apps = max_traffic.nlargest(3)

print("Top 3 Applications by Maximum Traffic:")
display(top_3_apps)


```

Task 2.1 - Part 4: Elbow Method to Optimize K for K-Means

```python
from sklearn.metrics import silhouette_score

def elbow_method(normalized_data):
    sse = []
    k_range = range(1, 11)
    
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(normalized_data)
        sse.append(kmeans.inertia_)  
    
   
    plt.figure(figsize=(8, 6))
    plt.plot(k_range, sse, 'bx-')
    plt.xlabel('k')
    plt.ylabel('SSE')
    plt.title('Elbow Method For Optimal k')
    plt.show()


elbow_method(normalized_metrics)


for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_metrics)
    score = silhouette_score(normalized_metrics, kmeans.labels_)
    print(f'Silhouette score for k={k}: {score}')
```
